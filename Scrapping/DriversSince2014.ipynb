```python
import pandas as pd
drivers=pd.read_csv("C:/Users/Youssef Tfifhq/Documents/Supports de cours , projets et examens/4ERP-BI/PBI/Formula 1/drivers.csv",header=0,sep =',')
drivers
```

```python
penalties=pd.read_csv("C:/Users/Youssef Tfifhq/Documents/Supports de cours , projets et examens/4ERP-BI/PBI/Formula 1/penalties2.csv",header=0,sep =',')
penalties
```

```python
penalties=penalties.rename(columns={"First": "forename", "last": "surname"})
```

```python
drivers1 = pd.merge(drivers, penalties, how="left", on=["forename", "surname"])
drivers1
```

```python
drivers1=drivers1.drop(['Driver', 'Event', 'Date', 'Points.1', 'Reason', 'driverCode'], axis = 1)
drivers1
```

```python
result=drivers1.drop_duplicates()
result
```

```python
result = result.drop(result[result.number=='\\N'].index)
result
```
